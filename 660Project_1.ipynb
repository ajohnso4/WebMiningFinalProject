```python
import pandas as pd

# Load the data
data = pd.read_csv("data.csv")

```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data["interview_title"], data["offer_acceptance"], test_size=0.3, random_state=42)

```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# Create a pipeline with CountVectorizer and LogisticRegression
pipe = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('classifier', LogisticRegression())])

```

```python
# Define the hyperparameters to tune
param_grid = {
    'vectorizer__ngram_range': [(1,1), (1,2), (2,2)],
    'classifier__C': [0.01, 0.1, 1, 10]
}

```

```python
# Perform grid search to find the best hyperparameters
grid = GridSearchCV(pipe, param_grid, cv=5)
grid.fit(X_train, y_train)

# Print the best hyperparameters and accuracy
print("Best hyperparameters: ", grid.best_params_)
print("Accuracy: ", grid.best_score_)

```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
# Create a pipeline with TfidfVectorizer and Naive Bayes
pipe = Pipeline([
    ('vectorizer', TfidfVectorizer()),
    ('classifier', MultinomialNB())])

```

```python
# Define the hyperparameters to tune
param_grid = {
    'vectorizer__ngram_range': [(1,1), (1,2), (2,2)],
    'classifier__alpha': [0.01, 0.1, 1, 10]
}

```

```python
# Perform grid search to find the best hyperparameters
grid = GridSearchCV(pipe, param_grid, cv=5)
grid.fit(X_train, y_train)

# Print the best hyperparameters and accuracy
print("Best hyperparameters: ", grid.best_params_)
print("Accuracy: ", grid.best_score_)
```
